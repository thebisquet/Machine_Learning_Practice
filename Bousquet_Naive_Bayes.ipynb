```python
import string
from nltk.corpus import stopwords

def temp_process(mess):
    STOPWORDS = stopwords.words('english')
    nopunc = [char for char in mess if char not in string.punctuation]
    nopunc = ''.join(nopunc)
    nopunc = ' '.join([word for word in nopunc.split() if word.lower() not in STOPWORDS])
    return nopunc
```

```python
sms['clean_msg'] = sms.text.apply(temp_process)
sms.head()
```

```python
words = sms.clean_msg.apply(lambda x: [word.lower() for word in x.split()])
```

```python
from collections import Counter
my_words = Counter()

for msg in words:
    my_words.update(msg)
    
print(my_words.most_common(50))
```

```python
X = sms.clean_msg
y = sms.target
print(X.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vect = CountVectorizer()
vect.fit(X_train)
X_train_dtm = vect.transform(X_train)
```

```python
X_test_dtm = vect.transform(X_test)
```

```python
X_train_dtm
```

```python
X_test_dtm
```

```python
from sklearn.naive_bayes import MultinomialNB
nb = MultinomialNB()
nb.fit(X_train_dtm, y_train)
```

```python
y_pred_class = nb.predict(X_test_dtm)
```

```python
from sklearn import metrics
metrics.accuracy_score(y_test, y_pred_class)
```

```python
metrics.confusion_matrix(y_test, y_pred_class)
```
