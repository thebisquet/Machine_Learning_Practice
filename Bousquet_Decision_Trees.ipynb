```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(cars)
```

```python
import wittgenstein as lw
ripper = lw.RIPPER(verbosity=0)
ripper.fit(train, class_feat='acceptability', pos_class='acc')
```

```python
ripper.ruleset_.out_pretty()
```

```python
X_test = test.drop('acceptability', axis=1)
y_test = test['acceptability']
```

```python
from sklearn.metrics import precision_score, recall_score
precision = ripper.score(X_test, y_test, precision_score)
recall = ripper.score(X_test, y_test, recall_score)
accuracy = ripper.score(X_test, y_test)
cond_count = ripper.ruleset_.count_conds()
print(f'precision: {precision} accuracy: {accuracy} recall: {recall} conds: {cond_count}')
```
