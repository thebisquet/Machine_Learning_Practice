```python
import numpy as np
import pandas as pd

data = pd.read_csv('../input/skin-cancer-mnist-ham10000/hmnist_28_28_L.csv')
data.info()
```

```python
data.describe()
```

```python
y = data.label
X = data.drop(columns=['label'])
```

```python
X_np = X.to_numpy()
print(X_np.shape)
print(X_np[0])
```

```python
from matplotlib import pyplot as plt
plt.gray()
plt.matshow(X_np[0].reshape(28,28))
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
standardizer = StandardScaler()
X_std = standardizer.fit_transform(X_np)
print(X.describe())
print(pd.DataFrame(X_std).describe())
```

```python
from sklearn.preprocessing import MinMaxScaler
minmaxer = MinMaxScaler()
X_nm = minmaxer.fit_transform(X_np)
print(pd.DataFrame(X_nm).describe())
```

```python
X_scaled = X_np / 255
print(pd.DataFrame(X_scaled).describe())
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_np, y, test_size=0.2, random_state=42)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
X_train_scaled = X_train / 255
X_test_scaled = X_test / 255
```

```python
from sklearn.neighbors import NearestNeighbors
knn = NearestNeighbors(n_neighbors=5).fit(X_train_scaled)
distances, indices = knn.kneighbors([X_test_scaled[0]])
print(indices)
print(y[indices.squeeze()])
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 10)
knn.fit(X_train, y_train)
knn.predict([X_test[0]])
```

```python
knn.score(X_test_scaled, y_test)
```

```python
from sklearn.model_selection import GridSearchCV
from math import sqrt
knn = KNeighborsClassifier()
search_space = [{'n_neighbors': list(range(1,len(X_train)//2))}]
grid_search_vals = GridSearchCV(knn, search_space).fit(X_train_scaled, y_train)
```

```python
grid_search_vals.best_estimator_.get_params()['n_neighbors']
```

```python
knn = KNeighborsClassifier(n_neighbors = grid_search_vals.best_estimator_.get_params()['n_neighbors'])
knn.fit(X_train_scaled, y_train)
knn.score(X_test_scaled, y_test)
```
